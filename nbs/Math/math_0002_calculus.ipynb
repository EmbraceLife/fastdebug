### logarithmic rules

```python
expr = log(x*y, a)
expr
```

```python
expr.factor()
```

```python
expr.subs({a:2, x:4, y:4})
expr.subs({a:2, x:4, y:4}).evalf(2)
```

```python
expr1 = (log(x, a) + log(y, a))
expr1
```

```python
expr1.factor()
```

```python
expr1.subs({a:2, x:4, y:4})
expr1.subs({a:2, x:4, y:4}).evalf(2)
```

```python
Eq(expr, expr1) # rule 1
```

```python
expr2 = log(x/y, a)
expr3 = log(x,a) - log(y,a)
Eq(expr2, expr3) # rule 2
```

```python
test_eq(expr2.subs({a:2, x:4, y:8}).evalf(), expr3.subs({a:2, x:4, y:8}).evalf())
```

```python
expr4 = log(x**y, a)
expr5 = y*log(x, a)
Eq(expr4, expr5) # rule 3
```

```python
test_eq(expr4.subs({a:2, x:4, y:8}).evalf(), expr5.subs({a:2, x:4, y:8}).evalf())
```

```python
log(1, a).evalf() # rule 4
Eq(log(1,a), 0)
```

```python
log(0, a).evalf() # rule 5, equal undefined
```

### finding domain of $\frac{sin(x)}{x}$ using continuous_domain

```python
from sympy import plot, symbols, sin, solve
```

```python
x = symbols('x')
expr = sin(x)/x
expr
```

```python
from sympy import Symbol, S
from sympy.calculus.util import continuous_domain
```

```python
S.Reals
```

```python
x = Symbol("x")
f = sin(x)/x
continuous_domain(f, x, S.Reals)
```
